```python
import spacy
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df['Category'].value_counts() #we found total 747 spam emails
```

```python
#Let's do encoding
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
x = df['Message']
y = df['spam']
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_cv = v.fit_transform(x_train.values)
x_train_np = x_train_cv.toarray()
```

```python
v.get_feature_names_out()[7454]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_cv,y_train)
```

```python
x_test_cv = v.transform(x_test)
y_pred = model.predict(x_test_cv)
```

```python
x_test.shape
```

```python
#checking score of model
model.score(x_test_cv,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#create sklearn pipelinee
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
clf.fit(x_train,y_train)
clf.predict(x_test)
```

```python
df2 = pd.read_csv('movies_sentiment_data.csv')
df2.head()
```

```python
df2['Category'] = df2['sentiment'].apply(lambda x: 1 if x=='positive' else 0)
df2.head()
```

```python
x = df2['review']
y = df2['Category']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c=  CountVectorizer()
x_train_cv = c.fit_transform(x_train.values)
x_train_np = x_train_cv.toarray()
x_train_np
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from  sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
#using Naive bayes
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
#using RandomForest
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('rb', RandomForestClassifier())
])
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
#using KNN
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('knn', KNeighborsClassifier())
])
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```
